```python
import pandas as pd
```

```python
dataset=pd.read_csv("project_risk_raw_dataset.csv")
dataset
```

```python
#Checking is null is available or not
dataset.isnull().sum()
```

```python
dataset["Tech_Environment_Stability"].fillna(0,inplace=True)
dataset["Change_Control_Maturity"].fillna(0,inplace=True)
dataset["Risk_Management_Maturity"].fillna(0,inplace=True)
```

```python
dataset.isnull().sum()
```

```python
from Univariate import Univariate
qual,quan=Univariate.quanQual(dataset)
quan
```

```python
descrptive=Univariate.IQR(quan,dataset)
descrptive
```

```python
lesser,greater=Univariate.findOutlier(quan,descrptive)
```

```python
greater
```

```python
replaceOutliers=Univariate.replaceOutlier(lesser,greater,dataset,descrptive)
replaceOutliers
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
```

```python
df
```

```python
df=pd.DataFrame(df,columns=quan)
df
```

```python
category=dataset[qual]
category
```

```python
frame=[df,category]
```

```python
preprocessed=pd.concat(frame,axis=1)
preprocessed
```

```python
preprocessed.isnull().sum()
```

```python
#Now our data is pre processed clearly 
#In this dataset we doesn't have any 'null' values and 'outliers'
preprocessed.to_csv("Preprocessed_project_risk.csv",index=False)
```
