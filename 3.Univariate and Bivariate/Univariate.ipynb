# Probability Density Function(PDF)

```python
#PDF-Probability Density Function
def get_pdf_probability(preprocessed,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(preprocessed,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = preprocessed
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob

get_pdf_probability(preprocessed["Stakeholder_Count"],5,20)
```

# Standard Noramal Distribution(SND)

```python
#SND-Standard normal ditribution

def stdNBgraph(preprocessed):
        # Coverted to standard Normal Distribution
    mean=preprocessed.mean()
    std=preprocessed.std()

    values=[i for i in preprocessed]

    z_score=[((j-mean)/std) for j in values]

    sns.distplot(z_score,kde=True)

    sum(z_score)/len(z_score)
    
stdNBgraph(preprocessed["Stakeholder_Count"])
```
